# **DATA** **VISUALIZATIONS**


```python
#Income vs. Monthly Expenses

import seaborn as sns
import matplotlib.pyplot as plt

data["Total_Expenses"] = data[[
    "Rent", "Loan_Repayment", "Groceries", "Transport",
    "Eating_Out", "Entertainment", "Utilities", "Healthcare"
]].sum(axis=1)

plt.figure(figsize=(10, 6))
sns.scatterplot(data, x="Income", y="Total_Expenses", hue="City_Tier", palette="viridis")
plt.title("Income vs Total Monthly Expenses by City Tier")
plt.xlabel("Income")
plt.ylabel("Total Expenses")
plt.show()
```

```python
#Boxplot of Income by City Tier

plt.figure(figsize=(10, 6))
sns.boxplot(data, x="City_Tier", y="Income", palette="pastel")
plt.title("Income Distribution by City Tier")
plt.xlabel("City Tier")
plt.ylabel("Income")
plt.show()
```

```python
#Histogram: Income Distribution

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(data["Income"], bins=30, color="skyblue", edgecolor="black")
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

```python
#Pie Chart: Expense Distribution

expense_columns = [
    "Rent", "Loan_Repayment", "Groceries", "Transport",
    "Eating_Out", "Entertainment", "Utilities", "Healthcare"
]

total_expenses = data[expense_columns].mean()
plt.figure(figsize=(8, 8))
plt.pie(total_expenses, labels=expense_columns, autopct="%1.1f%%", colors=plt.cm.Paired.colors)
plt.title("Average Expense Distribution")
plt.show()
```

```python
#Bar Plot: Average Spending by Category

plt.figure(figsize=(12, 6))
total_expenses.sort_values().plot(kind='bar', color="coral")
plt.title("Average Spending by Expense Category")
plt.xlabel("Expense Category")
plt.ylabel("Average Monthly Spending")
plt.show()
```
